### Performance Evaluation of Stacking Regressor on HOMO Energy Predictions (QM9 Dataset)

```python
import pandas as pd
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np

# Read CSV file
df = pd.read_csv(r"E:\Manuscript\StackingReg._homo_test_actual_vs_predicted_51_QM9.csv")

# Check column names
print(df.columns)

# Example: adjust column names if needed
y_true = df['Actual']
y_pred = df['Predicted']

# Calculate metrics
r2 = r2_score(y_true, y_pred)
mse = mean_squared_error(y_true, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_true, y_pred)

print(f"R²: {r2:.8f}")
print(f"MSE: {mse:.8f}")
print(f"RMSE: {rmse:.8f}")
print(f"MAE: {mae:.8f}")
```

### Compute Pearson and Spearman correlationsto confirm numerically (no trivial correlation)

```python
import pandas as pd
from scipy.stats import spearmanr
import numpy as np

# Load your dataset
file_path = r"E:\QM9_sci\HLP-Stack\data\processed\01\04_scaled_common_2D_3D_descriptors_QM9_61.csv\04_scaled_common_2D_3D_descriptors_QM9_61.csv"
df = pd.read_csv(file_path)

print(df.shape)
print(df.columns)

target_cols = ['homo', 'lumo']
feature_cols = [c for c in df.columns if c not in target_cols]

results = []

for t in target_cols:
    # Pearson correlation
    pearson_corr = df[feature_cols].corrwith(df[t], method='pearson')
    # Spearman correlation
    spearman_corr = df[feature_cols].apply(lambda x: spearmanr(x, df[t])[0])

    # Combine into DataFrame
    corr_df = pd.DataFrame({
        'Feature': feature_cols,
        f'Pearson_{t}': pearson_corr.values.round(4),
        f'Spearman_{t}': spearman_corr.values.round(4)
    })
    results.append(corr_df)

# Merge results for homo and lumo
corr_table = results[0].merge(results[1], on='Feature')
corr_table.set_index('Feature', inplace=True)

# Identify trivial correlations
trivial_corr = corr_table[(corr_table.abs() >= 0.95).any(axis=1)]

print("Features with potentially trivial correlation (|r| ≥ 0.95):")
if trivial_corr.empty:
    print("None found — no trivial correlations with HOMO or LUMO.")
else:
    print(trivial_corr)

# Sort all correlations by absolute Pearson values for clarity
corr_table_sorted = corr_table.reindex(
    corr_table['Pearson_homo'].abs().sort_values(ascending=False).index
)

# Save full, sorted correlation table
save_path = r"E:\QM9_sci\HLP-Stack\data\processed\01\corr_table_homo_lumo_sorted.csv"
corr_table_sorted.to_csv(save_path)

print(f"\nSorted correlation table saved successfully to:\n{save_path}")

# Show top 10 strongest correlations for inspection
print("\nTop 10 absolute correlations with HOMO:")
print(corr_table_sorted[['Pearson_homo', 'Spearman_homo']].abs().head(10))

print("\nTop 10 absolute correlations with LUMO:")
print(corr_table_sorted[['Pearson_lumo', 'Spearman_lumo']].abs().head(10))
```

### Feature Correlations with HOMO and LUMO Energies (Pearson & Spearman)

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select correlations to plot
corr_to_plot = corr_table[[f'Pearson_homo', f'Spearman_homo', f'Pearson_lumo', f'Spearman_lumo']]
corr_to_plot = corr_to_plot.loc[corr_to_plot.abs().mean(axis=1).sort_values(ascending=False).index]

plt.figure(figsize=(8, 10))

# ax = sns.heatmap(corr_to_plot, cmap='vlag', center=0, annot=corr_to_plot.abs() > 0.5, fmt=".2f")

ax = sns.heatmap(
    corr_to_plot,
    cmap='vlag',          # visually appealing diverging colormap
    center=0,
    linewidths=0.5,       # grid lines between cells
    linecolor='gray',
    annot=True, fmt=".3f",
    annot_kws={"size": 9, "weight": "bold"},
    cbar_kws={'label': 'Correlation'},
)

for pos in [2]:
    ax.axvline(pos, color='black', linewidth=1.2)

# Title
plt.title('Feature Correlations with HOMO and LUMO Energies (Pearson & Spearman)', 
          fontsize=12, fontweight='bold')

# Axis labels
plt.xlabel('Correlation Type', fontsize=10, fontweight='bold')
plt.ylabel('Features', fontsize=10, fontweight='bold')

# Axis ticks
plt.xticks(rotation=45, fontsize=10, fontweight='bold')
plt.yticks(rotation=0, fontsize=10, fontweight='bold')

# Colorbar customization
cbar = ax.collections[0].colorbar
cbar.set_label('Correlation', fontsize=10, fontweight='bold')      # label
cbar.ax.tick_params(labelsize=10)                                  # tick labels
cbar.ax.yaxis.label.set_weight('bold')

plt.tight_layout()

# Save figure at high resolution
save_path = r"E:\QM9_sci\HLP-Stack\data\processed\01\heatmap_homo_lumo_corr.png"
plt.savefig(save_path, dpi=600)
plt.show()

print(f"Heatmap saved at 600 dpi to:\n{save_path}")
```
